```python
from pydub import AudioSegment
import os
import datetime
import time

def get_today():
	now = time.localtime()
	s = "%04d-%02d-%02d" % (now.tm_year, now.tm_mon, now.tm_mday)
	return s

today = get_today()

ths_lists = []

for file in paths:
    filename, file_extension = os.path.splitext(file)
    if file_extension == ".mp3":
        dst = file.replace(".mp3","")+".wav"                                                             
        audSeg = AudioSegment.from_file(file, format="mp4")  # convert wav from mp3 
        audSeg.export(dst, format="wav")
        song = AudioSegment.from_wav(dst)
        sec = 10000     # 10초 변수
        block = song[sec * 8 : sec * 11]     # 1분 20초 ~ 1분 50초 구간만 사용하기 위한 변수(하이라이트 추출)
        path = general_path
        # inner_dir = "/converted/_30secs"
        # inner_dir = "/converted/" + today # 테스트 버전 1 // 컨버팅 후 저장 경로
        inner_dir = "/converted/" + today + "_" # 테스트 버전 2
        dir = path + inner_dir    
        if not os.path.exists(dir):
            os.makedirs(dir)
        name = f'{file.replace(".mp3","").replace(path,"")}_30.wav'
        full_path = path+inner_dir+name
        block.export(full_path, format='wav', parameters=["-q:a", "10", "-ac", "1"])
        print(f"Converting Successful // converted filename: {full_path}")
        ths_lists.append(full_path)
    else:
        print(file)
```
